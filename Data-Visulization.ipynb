```python
%matplotlib inline  
## magic ommandsimpoor

import numpy as np
```

```python
x=np.arange(1,20)
y= x**2 ## vectorization 
plt.title("Squares Vector")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.xlim(-20,20)
plt.ylim(1,300)

plt.plot(x,y,'yv',label='graph lines')
plt.legend()
```

```python
x=np.arange(1,20)
y= x**2 ## vectorization 
x2=np.arange(1,20)
y2=x2+50
x2=np.arange(1,20)
plt.title("Squares Vector")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.xlim(-20,20)
plt.ylim(1,300)

plt.plot(x,y,'yv',label='graph lines')
plt.plot(x2,y2,'ro',label='second graph lines')

plt.text(7.5,40,"Hello")
plt.grid()
plt.legend()
```

```python
plt.plot(x,y,'b--',x2,y2,'ro')
plt.show()
```
