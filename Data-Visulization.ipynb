sigmoid function 
f(x)= 1/ (1+e^x)

so we can write e^x ====> math.exp(x)

```python
import math 
import numpy  as np
sigmoid=lambda x: 1/(1+math.exp(-x))
x=np.arange(-100,100)
y= [sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```
