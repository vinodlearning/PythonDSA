```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data
np.random.seed(10)
data = np.concatenate([
    np.random.normal(100, 10, 200),  # Main distribution
    np.random.normal(150, 5, 10)     # Some outliers
])

# Create box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=True, patch_artist=True, labels=['Sample Data'])

# Add labels and title
plt.ylabel('Values')
plt.title('Box-Whisker Plot Example')
plt.grid(True, linestyle='--', alpha=0.7)

# Add annotations
plt.annotate('Outliers', xy=(1.05, 150), xytext=(1.2, 155),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Maximum (excluding outliers)', xy=(1, 125), xytext=(1.2, 130),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Q3 (75th percentile)', xy=(1, 110), xytext=(1.2, 115),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Median', xy=(1, 100), xytext=(1.2, 100),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Q1 (25th percentile)', xy=(1, 90), xytext=(1.2, 85),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Minimum (excluding outliers)', xy=(1, 75), xytext=(1.2, 70),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.savefig('box_plot_example.png')
plt.show()
```
